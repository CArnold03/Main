```python
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
#import urllib.request
#hand = urllib.request.urlopen('https://www.scripts.com/script/deadpool_1459')
deadpool=[]
wade=[]
mr_pool=[]
pool_boy=[]
def read_in_dataset(file_loc):
    names =  np.genfromtxt(file_loc, dtype = str, delimiter="")  
    return names
dataset_location = "datasets/Pool.txt"
name = read_in_dataset(dataset_location)
for line in name:
    #l = line.decode().strip()
    deadpools=re.findall('Deadpool', line)
    wades=re.findall('Wade', line)
    mr_pools=re.findall('Mr. Pool', line)
    pool_boys=re.findall('Pool Boy', line)
    deadpool.append(deadpools)
    wade.append(wades)
    mr_pool.append(mr_pools)
    pool_boy.append(pool_boys)
```

```python
dc=0
wc=0
mc=0
pc=0
for word in deadpool:
    dc+=1
for word in wade:
    wc+=1
for word in mr_pool:
    mc+=1
for word in pool_boy:
    pc+=1
```

```python
print("Deadpool: ", dc, "Is greater than: wade? ", dc>wc, "Is greater than: Mr Pool? ", dc>mc, "Is greater than: Pool Boy? ", dc>pc)
#print("Wade: ", wc, "Is greater than: Deadpool? ", dc<wc, "Is greater than: Mr Pool? ", wc>mc, "Is greater than: Pool Boy? ", wc>pc)
#print("Mr Pool: ", mc, "Is greater than: Deadpool? ", dc<mc, "Is greater than: Wade? ", mc>wc, "Is greater than: Pool Boy? ", mc>pc)
#print("Pool Boy: ", pc, "Is greater than: Deadpool? ", dc<pc, "Is greater than: Wade? ", pc>wc, "Is greater than: Mr Pool? ", mc<pc)
```
