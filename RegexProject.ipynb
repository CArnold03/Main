```python
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
import urllib.request
hand = urllib.request.urlopen('https://en.wikipedia.org/wiki/Hiram_Maxim')
dates=[]
for line in hand:
    l = line.decode().strip()
    date=re.findall('[1][8-9][0-9][0-9]', l)
    if date != []:
        dates.append(date)
print (dates)
```

```python
plt.hist(dates, bins= 1, orientation="horizontal")
plt.title("Maxim's Dates")
```
