```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_in_dataset(file_loc):
    num_clicks_per_minute =  np.genfromtxt(file_loc, dtype = int, delimiter=",")  
    return num_clicks_per_minute
```

```python
dataset_location = "datasets/geiger_counts_B.txt"
num_clicks = read_in_dataset(dataset_location)
mu=np.mean(num_clicks)
var=np.std(num_clicks)**2
```

```python
plt.hist(num_clicks, bins = 15)
plt.title('Average number of clicks per minute')
print('Average: ', mu)
print('Variance: ', var)
if mu==var:
    print('Are equal: True')
if mu!=var:
    print('Are equal: False')
```
