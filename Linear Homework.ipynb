```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy
import scipy.stats
```

```python
def read_in_dataset(file_loc):
  dataset =  np.genfromtxt(file_loc, dtype = int, delimiter=",", skip_header = 1)
  return dataset
dataset_location = "datasets/linear-trend-data.csv"
trend = read_in_dataset(dataset_location)
```

```python
def func(a, b):
    return a*b
```

```python
x=trend[:, 1]
y=trend[:, 0]
print ("x: ", x, "y: ", y)
```

```python
plt.figure() 
plt.scatter(x, y)
plt.xlabel("conductance")
plt.ylabel("1/tau")
plt.title("Before")
```

```python
params, params_covariance=scipy.optimize.curve_fit(func, x, y)
plt.figure()
plt.scatter(x, y, label='data')
plt.plot(x, func(x, params[0]), label='fit')
plt.xlim(xmax=1, xmin=-1)
plt.xlabel("conductance")
plt.ylabel("1/tau")
plt.title("After")
plt.legend(loc='best')
```

```python
print('y=%.2f*x+0' % (params[0]))
s=math.sqrt(params_covariance.reshape(-1)[0])
print("Lack of Certainty: ", s)
```
